### 学習モデルのパラメータの学習

> 想像上の動物が生涯を通じて食物を認識するにつれ、食物の典型的なサイズについての期待（vpおよびΣpによって記述される）や、光の強度を観察する際に犯す誤差の量（Σuによって記述される）についての期待を精緻化したいと考えるかもしれません。したがって、各刺激の後にパラメータvp、Σp、およびΣuを更新し、徐々にそれらを洗練させたいと考えるかもしれません。
我々は、知覚される光の強度uが最も驚くべきではない、言い換えれば最も予想されるモデルパラメータを選択したいと考えています。
