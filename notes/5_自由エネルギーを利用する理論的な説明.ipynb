# 自由エネルギーを利用する理論的な説明
自由エネルギー(-F)を以下のように定義して、自由エネルギーを最小化(Fの最大化)することで事後分布の近似式を得ることが  
できることを見てきたが、その理論的に説明する。

> 事後分布p(v|u)は複雑な形をしている可能性があるため、我々は別の分布、つまりq(v)でそれを近似することにします。重要なことに、q(v)が標準的な形をしていると仮定するため、この典型的な分布のパラメーターによってそれを特徴づけることができます。例えば、q(v)が正規分布であると仮定する場合、それを完全に説明するために、平均と分散の2つの数値を推定するだけで、任意の形の分布を特徴づけるために必要かもしれない無限の数の数値を推定する必要はありません。

p(v|u)とq(v)の近似は、Kullback-Leibler情報量を最小化することで実行する。
